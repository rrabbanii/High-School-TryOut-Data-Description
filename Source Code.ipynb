##**Number 1**##

Data Description and Box Plot Visualization without using an Interval Class

```python
# Importing Python Modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading csv file
df = pd.read_csv('/content/Data tryout AD Se KAb.csv')
# Shows the Info of the Data
df.info()
```

```python
# Remove 'no' from the Data Columns
df.drop('no', inplace=True, axis=1)

# Convert Data Type from object to float
df["Bhs. Ind"] = pd.to_numeric(df["Bhs. Ind"], errors = 'coerce')
df["Mtk"] = pd.to_numeric(df["Mtk"], errors = 'coerce')
df["Bhs. Ing"] = pd.to_numeric(df["Bhs. Ing"], errors = 'coerce')
df["IPA"] = pd.to_numeric(df["IPA"], errors = 'coerce')

# Shows Information and Description about the DataFrame including Minimum and Maximum Value | Mean and Median | Quartile 1 and Quartile 3
df.info()
df.describe()
```

```python
# Shows the Mode of the DataFrame
df.mode()
```

```python
# Shows the Interquartile Range of the DataFrame
Q1=df['Bhs. Ind'].quantile(0.25)
Q3=df['Bhs. Ind'].quantile(0.75)
IQR1 = Q3-Q1
print("Bhs. Ind Interquartile Range: ")
print(IQR1)

# Shows the Interquartile Range of the DataFrame
Q1=df['Mtk'].quantile(0.25)
Q3=df['Mtk'].quantile(0.75)
IQR2 = Q3-Q1
print("Mtk Interquartile Range: ")
print(IQR2)

# Shows the Interquartile Range of the DataFrame
Q1=df['Bhs. Ing'].quantile(0.25)
Q3=df['Bhs. Ing'].quantile(0.75)
IQR3 = Q3-Q1
print("Bhs. Ing Interquartile Range: ")
print(IQR3)

# Shows the Interquartile Range of the DataFrame
Q1=df['IPA'].quantile(0.25)
Q3=df['IPA'].quantile(0.75)
IQR4 = Q3-Q1
print("IPA Interquartile Range: ")
print(IQR4)
```

```python
# Shows the Variances of the DataFrame
df.var()
```

```python
# Converting DataFrame into a BoxPlot
df.boxplot(column=['Bhs. Ind', 'Mtk', 'Bhs. Ing', 'IPA', 'rata'])
```
